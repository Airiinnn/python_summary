# MongoDB (Local Only)
- Creating and inserting into a collection
- Retrieving documents

```python
# Creating and inserting into a collection

from pymongo import MongoClient

client = MongoClient()
# client = MongoClient('localhost', 27017)
# client = MongoClient('mongodb://localhost:27017')

db = client.mydatabase
# db = client["mydatabase"]

students = db.students

student1 = {
    "name" : "Alex Liswandy",
    "gender" : "M",
    "age" : 16
}

students.insert_one(student1)

student2 = {
    "name" : "Bob Bee",
    "gender" : "M",
    "age" : 17
}

student3 = {
    "name" : "Charlie Lee",
    "gender" : "M",
    "age" : 13
}

students.insert_many([student2, student3])
```

```python
# Retrieving documents

from pymongo import MongoClient

client = MongoClient()
# client = MongoClient('localhost', 27017)
# client = MongoClient('mongodb://localhost:27017')

db = client.mydatabase
# db = client["mydatabase"]

student = students.find_one({"name": "Alex Liswandy"})
print(student)
```
