```python
# Linked list (unordered)

class Node:
    
    def __init__(self, data, link=None):
        self.data = data
        self.link = link
        
        

class LinkedList:
    
    def __init__(self):
        self.head = None
        
    def add(self, data):
        node = Node(data)
        node.link = self.head
        self.head = node
        
    def update(self, target, data):
        curr = self.head
        found = False
        
        while curr != None and not found:
            if curr.data == target:
                curr.data = data
                found = True
                
            else:
                curr = curr.link
                
        if found:
            print("Updated")
            
        else:
            print("Not found")
            
    def delete(self, target):
        if self.head != None:
            if self.head.data == target:
                self.head = self.head.link
                print("Deleted")
                
            else:
                curr = self.head
                prev = self.head
                found = False

                while curr != None and not found:
                    if curr.data == target:
                        prev.link = curr.link
                        found = True

                    else:
                        prev = curr
                        curr = curr.link

                if found:
                    print("Deleted")
                else:
                    print("Not found")
            
    def search(self, target):
        curr = self.head
        found = False
        
        while curr != None and not found:
            if curr.data == target:
                found = True
                
            else:
                curr = curr.link
                
        if found:
            print("Found")
        else:
            print("Not found")
            
    def size(self):
        count = 0
        curr = self.head
        
        while curr != None:
            count += 1
            curr = curr.link
            
        return count
        
    def display(self):
        curr = self.head
        
        while curr != None:
            print(curr.data, end=' ')
            curr = curr.link
            
        print()
        
        
        
# main
ll = LinkedList()
ll.add(1)
ll.add(2)
ll.add(3)
ll.add(9)
ll.add(8)
ll.add(7)
print(ll.size())
ll.display()
ll.update(2, 4)
ll.update(5, 10)
print(ll.size())
ll.display()
ll.delete(7)
ll.delete(10)
print(ll.size())
ll.display()
```

```python
# Stack (array)

class Stack:
    
    MAX = 5
    
    def __init__(self):
        self.stack = [0 for i in range(self.MAX)]
        self.top = -1
        
    def isEmpty(self):
        return self.top == -1
    
    def isFull(self):
        return self.top == self.MAX - 1
    
    def push(self, data):
        if self.isFull():
            print("Cannot push to a full stack")
        
        else:
            self.top += 1
            self.stack[self.top] = data
            print("Added")
            
    def pop(self):
        if self.isEmpty():
            print("Cannot pop from empty stack")
            
        else:
            data = self.stack[self.top]
            self.top -= 1
            return data
        
    def peek(self):
        if self.isEmpty():
            print("Empty")
            
        else:
            print(self.stack[self.top])
        
    def display(self):
        if self.isEmpty():
            print("Empty")
            
        else:
            cur = self.top
            while cur > -1:
                print(self.stack[cur], end=' ')
                cur -= 1
                
            print()



# main
stack = Stack()
stack.peek()
stack.display()
stack.pop()
stack.push(1)
stack.push(3)
stack.push(2)
stack.push(5)
stack.push(6)
stack.push(4)
stack.peek()
stack.display()
stack.pop()
stack.peek()
stack.display()
```

```python
# Queue (array)

class Queue:
    
    MAX = 5
    
    def __init__(self):
        self.queue = [0 for i in range(self.MAX)]
        self.head = 0
        self.tail = 0
        
    def isEmpty(self):
        return self.head == self.tail
    
    def isFull(self):
        return self.size() == self.MAX - 1
    
    def size(self):
        if self.isEmpty():
            return 0
        
        elif self.tail > self.head:
            return self.tail - self.head
        
        else:
            return self.tail + self.MAX - self.head
        
    def enqueue(self, data):
        if self.isFull():
            print("Cannot enqueue to full queue")
            
        else:
            self.queue[self.tail] = data
            self.tail = (self.tail + 1) % self.MAX
            
    def dequeue(self):
        if self.isEmpty():
            print("Cannot dequeue from empty queue")
        
        else:
            data = self.queue[self.head]
            self.head = (self.head + 1) % self.MAX
            return data
        
    def peek(self):
        if self.isEmpty():
            print("Empty")
            
        else:
            print(self.queue[self.head])
            
    def display(self):
        if self.isEmpty():
            print("Empty")
            
        elif self.tail > self.head:
            i = self.head
            while i < self.tail:
                print(self.queue[i], end=' ')
                i += 1
                
            print()
                    
        else:
            i = self.head
            while i < self.MAX:
                print(self.queue[i], end=' ')
                i += 1
            
            i = 0
            while i < self.tail:
                print(self.queue[i], end=' ')
                i += 1
            
            print()
            
            
            
# main
q = Queue()
print(q.size())
q.peek()
q.display()
q.dequeue()
q.enqueue(1)
q.enqueue(3)
q.enqueue(2)
q.enqueue(5)
q.enqueue(4)
print(q.size())
q.peek()
q.display()
q.dequeue()
q.dequeue()
q.enqueue(1)
q.enqueue(3)
print(q.size())
q.peek()
q.display()
```

```python
# Queue (linked list)

class Node:
    
    def __init__(self, data, link=None):
        self.data = data
        self.link = link
        
        

class Queue:
    
    def __init__(self):
        self.head = None
        self.rear = None
        
    def isEmpty(self):
        return self.head == self.rear == None
    
    def enqueue(self, data):
        node = Node(data)
        
        if self.isEmpty():
            self.head = node
            
        else:
            self.rear.link = node
        
        self.rear = node
        
    def dequeue(self):
        if self.isEmpty():
            print("Cannot dequeue from empty queue")
            
        else:
            data = self.head.data
            
            if self.head.link == None:
                self.head = None
                self.rear = None
                
            else:
                self.head = self.head.link
                
            return data
        
    def peek(self):
        if self.isEmpty():
            print("Empty")
            
        else:
            print(self.head.data)
            
    def display(self):
        if self.isEmpty():
            print("Empty")
        
        else:
            curr = self.head

            while curr != None:
                print(curr.data, end=' ')
                curr = curr.link
                
            print()
                
                
                
# main
q = Queue()
q.peek()
q.display()
q.dequeue()
q.enqueue(1)
q.enqueue(3)
q.enqueue(2)
q.enqueue(5)
q.enqueue(4)
q.peek()
q.display()
q.dequeue()
q.dequeue()
q.enqueue(1)
q.enqueue(3)
q.peek()
q.display()
```

```python
# BST (linked list)

class BST:
    
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
    def insert(self, data):
        if data < self.data:
            if self.left == None:
                self.left = BST(data)
            else:
                self.left.insert(data)
                
        else: # data > self.data
            if self.right == None:
                self.right = BST(data)
            else:
                self.right.insert(data)
                
    def search(self, target):
        if self.data == target:
            return "Found"
        
        elif target < self.data:
            if self.left == None:
                return "Not found"
            else:
                return self.left.search(target)
                
        else: # target > self.data
            if self.right == None:
                return "Not found"
            else:
                return self.right.search(target)
                
    def lookup(self, target, parent=None):
        if self.data == target:
            return self, parent
        
        elif target < self.data:
            if self.left == None:
                return None, None
            else:
                return self.left.lookup(target, self)
                
        else: # target > self.data
            if self.right == None:
                return None, None
            else:
                return self.right.lookup(target, self)
                
    def delete(self, target):
        node, parent = self.lookup(target)
        
        if node == None:
            print("Not found")
            
        else:
            if node.left == node.right == None: # 0 child
                if parent == None:
                    node = None
                
                else:
                    if parent.left == node:
                        parent.left = None
                    else: # parent.right == node
                        parent.right = None
                    
            elif (node.left == None) != (node.right == None): # 1 child
                if node.left:
                    child = node.left
                else: # node.right
                    child = node.right
                    
                if parent == None:
                    node = child
                
                else:
                    if parent.left == node:
                        parent.left = child
                    else: # parent.right == node
                        parent.right = child
                    
            else: # 2 children
                temp = None
                child = node.left
                while child.right:
                    temp = child
                    child = child.right
                    
                node.data = child.data
                
                if temp:
                    temp.right = child.left
                else:
                    node.left = child.left
                
    def minimum(self):
        if self.left == None:
            print(f"Minimum: {self.data}")
        else:
            self.left.minimum()
            
    def maximum(self):
        if self.right == None:
            print(f"Maximum: {self.data}")
        else:
            self.right.maximum()
            
    def inorder(self):
        if self.left:
            self.left.inorder()
        print(self.data, end=' ')
        if self.right:
            self.right.inorder()
            
    def preorder(self):
        print(self.data, end=' ')
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
            
    def postorder(self):
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.data, end=' ')
        
    def reverse(self):
        if self.right:
            self.right.reverse()
        print(self.data, end=' ')
        if self.left:
            self.left.reverse()
            
            
            
# main
bst = BST(50)
bst.insert(30)
bst.insert(80)
bst.insert(10)
bst.insert(40)
bst.insert(70)
bst.insert(90)
bst.insert(60)
bst.insert(75)
bst.insert(73)
bst.inorder()
print()
bst.preorder()
print()
bst.postorder()
print()
bst.reverse()
print()
bst.delete(90) # 0 child
bst.delete(70) # 1 child
bst.delete(50) # 2 children
bst.delete(80) # 2 children
print("after")
bst.inorder()
print()
bst.preorder()
print()
bst.postorder()
print()
bst.reverse()
print()
print(bst.search(80))
print(bst.search(100))
bst.minimum()
bst.maximum()
```

```python
# Linear search

def linearSearch(arr, target):
    n = len(arr)
    i = 0
    
    while i < n:
        if arr[i] == target:
            return f"{target} found at index {i}"
        
        i += 1
        
    return "Not found"



# main
arr = [2, 4, 1, 5, 3]
print(linearSearch(arr, 5))
print(linearSearch(arr, 6))
```

```python
# Binary search (recursive)

def binarySearch(arr, target):
    if arr == []:
        return "Not found"
    
    mid = len(arr) // 2
    
    if target == arr[mid]:
        return "Found"
    elif target < arr[mid]:
        return binarySearch(arr[:mid], target)
    else:
        return binarySearch(arr[mid+1:], target)
        
        
        
# main
arr = [1, 2, 3, 4, 5, 6, 7, 8]
print(binarySearch(arr, 4))
print(binarySearch(arr, 11))
```

```python
# Binary search (iterative)

def binarySearch(arr, target):
    low = 0
    high = len(arr) - 1
    
    while high >= low:
        mid = (high + low) // 2
        
        if target == arr[mid]:
            return "Found"
        elif target < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1
            
    return "Not found"



# main
arr = [1, 2, 3, 4, 5, 6, 7, 8]
print(binarySearch(arr, 4))
print(binarySearch(arr, 11))
```

```python
# Bubble sort

def bubbleSort(arr):
    n = len(arr)
    swap = True
    
    while swap:
        swap = False
        
        for i in range(n-1):
            if arr[i] > arr[i+1]:
                temp = arr[i]
                arr[i] = arr[i+1]
                arr[i+1] = temp
                swap = True
                
        n -= 1
    
    return arr



# main
arr = [1, 7, 5, 4, 6, 2, 3, 9, 8]
print(bubbleSort(arr))
```

```python
# Insertion sort

def insertionSort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        
        while key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
            
        arr[j+1] = key
        
    return arr



# main
arr = [1, 7, 5, 4, 6, 2, 3, 9, 8]
print(insertionSort(arr))
```

```python
# Quicksort

def quicksort(arr):
    if arr == []:
        return []
    
    else:
        pivot = arr[0]
        left = []
        right = []
        
        for i in arr[1:]:
            if i < pivot:
                left.append(i)
            else:
                right.append(i)
                
        left = quicksort(left)
        right = quicksort(right)
        
        return left + [pivot] + right
    

    
# main
arr = [1, 7, 5, 4, 6, 2, 3, 9, 8]
print(quicksort(arr))
```

```python
# Quicksort (short)

def quicksort(arr):
    if arr == []:
        return []
    
    else:
        return quicksort([i for i in arr[1:] if i < arr[0]]) + [arr[0]] + quicksort([i for i in arr[1:] if i >= arr[0]])
    

    
# main
arr = [1, 7, 5, 4, 6, 2, 3, 9, 8]
print(quicksort(arr))
```

```python
# Merge sort

def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left = arr[:mid]
        right = arr[mid:]
        
        mergeSort(left)
        mergeSort(right)
        
        i = j = k = 0
        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                arr[k] = left[i]
                i += 1
                
            else:
                arr[k] = right[j]
                j += 1
                
            k += 1
            
        while i < len(left):
            arr[k] = left[i]
            i += 1
            k += 1
            
        while j < len(right):
            arr[k] = right[j]
            j += 1
            k += 1
            
    return arr
    

    
# main
arr = [1, 7, 5, 4, 6, 2, 3, 9, 8]
print(mergeSort(arr))
```
